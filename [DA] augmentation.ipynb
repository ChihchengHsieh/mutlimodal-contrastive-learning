```python
from torchvision.io import read_image
import torchvision.transforms as T
```

```python
aug_transform = T.Compose(
                [
                    T.RandomResizedCrop(
                        [351, 505], scale=(0.2, 1.0)),
                    T.RandomHorizontalFlip(p=0.5),
                    T.RandomRotation(15),
                    T.ColorJitter(contrast=0.5, saturation=0.5, brightness=.5, hue=.3),
                    T.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
                    T.ToPILImage(),
                ]
            )

to_pil = T.ToPILImage()
```

```python
xray = read_image('./ped_examples/4.png')
```

```python
aug_transform(xray)
```

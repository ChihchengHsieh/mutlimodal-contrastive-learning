```python
import torch
```

```python
a = torch.randn(16, 49, 32)
b = torch.randn(16, 49, 32)
c = torch.randn(16, 49, 32)
```

```python
r1 = (a @ b.permute(0, 2, 1)) @ c
```

```python
r2 = torch.einsum("bld,bdl,bld->bld", a, b.permute(0, 2, 1), c)
```

```python
r3 = torch.einsum("bij,bjd->bid", torch.einsum("bid,bdj->bij", a, b.permute(0, 2, 1)) , c)
```
